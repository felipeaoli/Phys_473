```python
import numpy as np 
import matplotlib.pyplot as plt 

z, hz, sig = np.loadtxt("./Hz_cosmicchron.txt").T
```

```python
plt.errorbar(z, hz, yerr=sig, fmt='.', capsize=3)
```

```python
def h(z, H0, om):
    return H0*np.sqrt(om*(1+z)**3 + 1-om )
```

```python
chi2_func = lambda z, H0, om: np.sum((h(z, H0, om) - hz)**2/sig**2)

harr = np.linspace(60, 90, 100)
# harr = np.linspace(60, 90, 100)

chi2 = np.array([chi2_func(z, hi, 0.3) for hi in harr])
Dchi2 = chi2-chi2.min()
plt.plot(harr, np.exp(-Dchi2/2))

hmin = harr[np.argmin(Dchi2)]

```
